```python
import numpy as np 
import matplotlib.pyplot as plt
import json
```

```python
# loads in data from Jsons 

with open('degree_entropy_examples') as f:
    degree_entropy = json.load(f)
with open('edge_entropy_examples') as f:
    edge_entropy = json.load(f)
with open('resistance_entropy_lloyds_trends') as f:
    trends = json.load(f)
with open('resistance_edge_degree') as f:
    all = json.load(f)
```

```python
#partitions data from jsons and graphs 

degree_entropies_300_std = np.array(trends["300 nodes"]["degree entropy"]["std"])
degree_entropies_200_std = np.array(trends["200 nodes"]["degree entropy"]["std"])
degree_entropies_100_std = np.array(trends["100 nodes"]["degree entropy"]["std"])

degree_entropies_300_mean = np.array(trends["300 nodes"]["degree entropy"]["mean"])
degree_entropies_200_mean = np.array(trends["200 nodes"]["degree entropy"]["mean"])
degree_entropies_100_mean = np.array(trends["100 nodes"]["degree entropy"]["mean"])


plt.plot(degree_entropies_300_mean,label="300 node graphs")
plt.fill_between(range(len(degree_entropies_300_mean)), (degree_entropies_300_mean - degree_entropies_300_std), (degree_entropies_300_mean + degree_entropies_300_std), color = 'b', alpha = .1)
plt.plot(degree_entropies_200_mean,label="200 node graphs")
plt.fill_between(range(len(degree_entropies_200_mean)), (degree_entropies_200_mean - degree_entropies_200_std), (degree_entropies_200_mean + degree_entropies_200_std), color = 'orange', alpha = .1)
plt.plot(degree_entropies_100_mean,label="100 node graphs")
plt.fill_between(range(len(degree_entropies_100_mean)), (degree_entropies_100_mean - degree_entropies_100_std), (degree_entropies_100_mean + degree_entropies_100_std), color = 'g', alpha = .1)
plt.xlabel("Lloyd Itteration")
plt.ylabel("Node Entropy")
plt.title("Node Entropy Per Lloyds Itteration")
plt.legend(bbox_to_anchor = (1.40, .9), loc='center right')
plt.show()

#multiplying by a scaling factor of ln(N) and graphing

degree_entropies_300_std = np.log(300)*np.array(trends["300 nodes"]["degree entropy"]["std"])
degree_entropies_200_std = np.log(200)*np.array(trends["200 nodes"]["degree entropy"]["std"])
degree_entropies_100_std = np.log(100)*np.array(trends["100 nodes"]["degree entropy"]["std"])

degree_entropies_300_mean = np.log(300)*np.array(trends["300 nodes"]["degree entropy"]["mean"])
degree_entropies_200_mean = np.log(200)*np.array(trends["200 nodes"]["degree entropy"]["mean"])
degree_entropies_100_mean = np.log(100)*np.array(trends["100 nodes"]["degree entropy"]["mean"])


plt.plot(degree_entropies_300_mean,label="300 node graphs")
plt.fill_between(range(len(degree_entropies_300_mean)), (degree_entropies_300_mean - degree_entropies_300_std), (degree_entropies_300_mean + degree_entropies_300_std), color = 'b', alpha = .1)
plt.plot(degree_entropies_200_mean,label="200 node graphs")
plt.fill_between(range(len(degree_entropies_200_mean)), (degree_entropies_200_mean - degree_entropies_200_std), (degree_entropies_200_mean + degree_entropies_200_std), color = 'orange', alpha = .1)
plt.plot(degree_entropies_100_mean,label="100 node graphs")
plt.fill_between(range(len(degree_entropies_100_mean)), (degree_entropies_100_mean - degree_entropies_100_std), (degree_entropies_100_mean + degree_entropies_100_std), color = 'g', alpha = .1)
plt.xlabel("Lloyd Itteration")
plt.ylabel("Node Entropy times $\ln(N)$")
plt.title("Node Entropy times $\ln(N)$ Per Lloyds Itteration")
plt.legend(bbox_to_anchor = (1.40, .9), loc='center right')
plt.show()
```
